```python
class Student:
  def __init__(self, name, marks):
        self.name = name
        self.marks = marks

  def total(self):
    print("Total:",sum(self.marks))
  def average(self):
    print("Average:",sum(self.marks)/len(self.marks))
  def high(self):
    print("Highest Mark:",max(self.marks))
  def check(self):
    if any(i < 35 for i in self.marks):
      print("Fail")
    else:
      print("Pass")
name="Emman"
marks=[90,80,95,23,94]
s=Student(name,marks)
s.total()
s.average()
s.high()
s.check()
```

```python
class Student:
  def __init__(self, name, scores):
        self.name = name
        self.scores = scores
  def average(self):
    print("Average Score:",sum(self.scores)/len(self.scores))
  def high(self):
    print("Best score:",max(self.scores))
  def check(self):
    count=0
    for i in self.scores:
      if i>=80:
        count+=1
    print("Months above 80: ",count)
name="Deva"
scores=[75, 82, 90, 65, 88, 70]
s=Student(name,scores)
s.average()
s.high()
s.check()
```

```python
class Vehicle:
  def __init__(self, vehicles):
        self.vehicles = vehicles
  def dup(self):
    dup=[]
    for i in self.vehicles:
      if self.vehicles.count(i)>1:
        dup.append(i)
    print("Duplicate Vehicles:",list(set(dup)))
  def un(self):
    a=list(set(self.vehicles))
    print("Unique Vehicles:",len(a))


vehicles=[101, 105, 110, 101, 115, 105]
s=Vehicle(vehicles)
s.dup()
s.un()
```

```python
class Cricket:
  def __init__(self, player, scores):
        self.player = player
        self.scores = scores
  def average(self):
    print("Average Score:",sum(self.scores)/len(self.scores))
  def low(self):
    print("Lowest score:",min(self.scores))
  def high(self):
    print("Highest score:",max(self.scores))
  def check(self):
    count=0
    for i in self.scores:
      if i>=50:
        count+=1
    print("Half Centuries: ",count)
player="Virat Kohli"
scores=[45, 102, 78, 35, 61]
s=Cricket(player,scores)
s.average()
s.high()
s.low()
s.check()
```

```python
from functools import reduce
class Ecommerce:
  def __init__(self,prices):
        self.prices = prices
  def total(self):
    a=reduce(lambda x,y:x+y,self.prices)
    print("Total Bill:",a)
  def cost(self):
    print("Costliest Item",max(self.prices))
  def dis(self):
    a=reduce(lambda x,y:x+y,self.prices)
    if(a>5000):
      b=0.85*a
      print("Discounted Bill:",b)
s=Ecommerce([1200, 3500, 800, 1500])
s.total()
s.cost()
s.dis()

```

```python
class Patient:
  def __init__(self,patient,temp):
    self.patient=patient
    self.temp=temp
  def high(self):
    print("Highest Temperature:",max(self.temp))
  def low(self):
    print("Lowest Temperature:",min(self.temp))
  def ab(self):
    count=0
    for i in self.temp:
      if i>100:
        count+=1
    print("Abnormal Readings:",count)
patient=input()
temp=list(map(float,input().split()))
p=Patient(patient,temp)
p.low()
p.high()
p.ab()
```

```python
class MovieTicket:
  def __init__(self,seats):
    self.seats=seats
  def book(self):
    print("Total Booked Seats:",len(self.seats))
    n=20
    available=[]
    for i in range(1,n+1):
      if i not in self.seats:
        available.append(i)
    print("Available Seats:",available)
seats=list(map(int,input().split()))
p=MovieTicket(seats)
p.book()
```

```python
class Student:
  def __init__(self,att):
    self.att=att
  def high(self):
    print("Highest Attendance:",max(self.att))
  def low(self):
    print("Lowest Attendance:",min(self.att))
  def eli(self):
    count=0
    for i in self.att:
      if i>=75:
        count+=1
    print("Eligible Students:",count)
s=Student([85, 70, 92, 65, 78, 88])
s.high()
s.low()
s.eli()
```
