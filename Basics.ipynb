```python
s1 = input()

s2 = s1[::-1]

list1 = []
list2 = []

for i in range(len(s1)):
    for j in range(i + 1, len(s1) + 1):
        sub1 = s1[i:j]
        list1.append(sub1)
for i in range(len(s2)):
    for j in range(i + 1, len(s2) + 1):
        sub2 = s2[i:j]
        list2.append(sub2)

a = set(list1)
b = set(list2)

common = a.intersection(b)
print(common)

ans = ""

for x in common:

    if x == x[::-1]:

        if len(x) > len(ans):
            ans = x

print(ans)
```

```python
s=input()

s1=s.lower()

l=['a','e','i','o','u']

v=[]
c=[]

for i in range(len(s1)):

    if s1[i] in l:
        v.append(s1[i])

    else:
        c.append(s1[i])

v = v[::-1]

ans = ""

m = min(len(v), len(c))

for i in range(m-1):

    ans += v[i]
    ans += c[i]

ans += v[-2]
ans += v[-1]
ans += c[-1]

print(ans)
```

```python
class Teacher:
  def func1(self):
    print("Teacher")
class Student(Teacher):
  def func2(self):
    print("Student")
class Parent(Student):
  def func3(self):
    print("Class C")
obj=Parent()
obj.func1()
obj.func2()
obj.func3()
```

```python
class Base1:
  def func1(self):
    print("Teacher")
class Base2:
  def func2(self):
    print("Student")
class Child(Base1,Base2):
  def func3(self):
    print("Class C")
obj=Child()
obj.func1()
obj.func2()
obj.func3()
```
